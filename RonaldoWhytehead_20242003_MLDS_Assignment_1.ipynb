Using list comprehension, generating a new list that has values for each: TransactionNumber modulo
4 Insert the list as a new column between CustomerName and TransactionNumber in the existing
DataFrame. The column name should be called TransactionHash.


```python
# Changing the data type name of TransactionNumber to be a integer
Sales_of_Customer_df['TransactionNumber'] = Sales_of_Customer_df['TransactionNumber'].astype(int)
# Creating Transaction Hash and making values modulo 4
transaction_hashes = [num % 4 for num in Sales_of_Customer_df['TransactionNumber']]
# Creation the Transaction Number Index
transaction_number_index = Sales_of_Customer_df.columns.get_loc('TransactionNumber')
# Inserting the Transaction Hash column into the DataFrame
Sales_of_Customer_df.insert(transaction_number_index, 'TransactionHash', transaction_hashes)
```

Create a new column 'AgeGroup' that categorizes customers into 'Youth' (below 30), 'Adult'
(30-60), and 'Senior' (above 60).

```python
# Creating a new column 'AgeGroup' which categorizes into 'Youth' (below 30), 'Adult' (30-60), and 'Senior (above 60)
bins = [0, 30, 60, float('inf')]
labels = ['Youth', 'Adult', 'Senior']
Sales_of_Customer_df['AgeGroup'] = pd.cut(Sales_of_Customer_df['Age'], bins=bins, labels=labels)
print("DataFrame reflecting AgeGroup Column")
print(Sales_of_Customer_df)
```
